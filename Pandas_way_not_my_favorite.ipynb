```python
import pandas as pd
```

```python
co_name = "Done A & B Construction"
fname = co_name + ".csv" # build file name from company name (special chars an issue in github)

url = "https://raw.githubusercontent.com/dsco/matt-contacts/main/Done%20A%20%26%20B%20Construction.csv"
```

```python
df = pd.read_csv(url, header=None, on_bad_lines='skip')
```

```python
df.head(10)
```

```python
# features used to strip out meaningless rows
contacts_flag = False
contacts_list = []
```

```python
# loop through the initial file and strip out everything
#but the 'contacts' rows
for index, row in df.iterrows():
        if row[0] == 'Contacts': # start reading the csv
            contacts_flag = True
            continue
        if row[0] == 'Projects': # stop reading the csv
            contacts_flag = False
            continue
        if len(row) != 0 and contacts_flag:
            contacts_list.append(row)
            
df2 = pd.DataFrame(contacts_list)
```

```python
df2.head(50) # a test view
```

```python
# Build a new dict for each contact and
# Build a dataframe of all contacts extracted from file
contact_id = 0
contacts = []
contact = {}
for index, row in df2.iterrows():
    if row[0] == 'Name':
        contact_id += 1
        contacts.append(contact)
        contact = {}
        
        contact['Company'] = co_name
        contact['Name'] = row[1]
        # contact['id'] = contact_id
    elif isinstance(row[0], str):
        contact[row[0]] = row[1]
df3 = pd.DataFrame(contacts)
```

```python
# clean up the edges
df3 = df3.drop(index=0)
del df3['Fax']
```
